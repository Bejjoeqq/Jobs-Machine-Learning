```python
#train
df_id = df.drop('id', axis=1)
df_train = df_id.drop('Gaji', axis=1)
df_target = df_id["Gaji"]

#test
df_test = df1.drop('id', axis=1)
```

```python
#ord
dictt = {
    "Tidak Pernah Bekerja":0,
    "Tanpa di Bayar":0,
    "Pemerintah Provinsi":4,
    "Pekerja Bebas Perusahaan":5,
    "Pemerintah Negara":3,
    "?":0,
    "Pemerintah Lokal":2,
    "Pekerja Bebas Bukan Perusahan":2,
    "Wiraswasta":1
}
dictt2 = {
    "SMA":1,
    "Pendidikan Tinggi":2,
    "Sarjana":3,
    "Master":5,
    "D4":2,
    "11th":0,
    "D3":2,
    "10th":0,
    "7th-8th":0,
    "Sekolah Professional":5,
    "9th":0,
    "12th":0,
    "Doktor":5,
    "5th-6th":0,
    "1st-4th":0,
    "SD":0
}
dictt3 = {
    "Menikah":4,
    "Belum Pernah Menikah":0,
    "Cerai":1,
    "Berpisah":0,
    "Janda":0,
    "Menikah LDR":1
}
dictt4 = {
    "Spesialis":4,
    "Perbaikan Kerajinan":1,
    "Ekesekutif Managerial":5,
    "Pemuka Agama":1,
    "Sales":2,
    "Servis Lainnya":0,
    "Mesin Inspeksi":1,
    "?":0,
    "Supir":1,
    "Pembersih":0,
    "Petani":1,
    "Tech-support":2,
    "Penjaga":2,
    "Asisten Rumah Tangga":0,
    "Tentara":3
}
dictt5 = {
    "Laki2":2,
    "Perempuan":1
}

df_train["Kelas Pekerja"] = df_train["Kelas Pekerja"].replace(dictt)
df_test["Kelas Pekerja"] = df_test["Kelas Pekerja"].replace(dictt)

df_train["Pendidikan"] = df_train["Pendidikan"].replace(dictt2)
df_test["Pendidikan"] = df_test["Pendidikan"].replace(dictt2)

df_train["Status Perkawinan"] = df_train["Status Perkawinan"].replace(dictt3)
df_test["Status Perkawinan"] = df_test["Status Perkawinan"].replace(dictt3)

df_train["Pekerjaan"] = df_train["Pekerjaan"].replace(dictt4)
df_test["Pekerjaan"] = df_test["Pekerjaan"].replace(dictt4)

df_train["Jenis Kelamin"] = df_train["Jenis Kelamin"].replace(dictt5)
df_test["Jenis Kelamin"] = df_test["Jenis Kelamin"].replace(dictt5)
```

```python
keuntungan1 =[x if x>0 else 0 for x in df_train["Keuntungan Kapital"]]
kerugian1 =[-x if x>0 else 0 for x in df_train["Kerugian Capital"]]
kapitaltotal1 = [x + y for x, y in zip(keuntungan1, kerugian1)]

keuntungan2 =[x if x>0 else 0 for x in df_test["Keuntungan Kapital"]]
kerugian2 =[-x if x>0 else 0 for x in df_test["Kerugian Capital"]]
kapitaltotal2 = [x + y for x, y in zip(keuntungan2, kerugian2)]
```

```python
df_train = df_train.drop('Keuntungan Kapital', axis=1)
df_train = df_train.drop('Kerugian Capital', axis=1)

df_test = df_test.drop('Keuntungan Kapital', axis=1)
df_test = df_test.drop('Kerugian Capital', axis=1)

df_train["Kapital"] = kapitaltotal1

df_test["Kapital"] = kapitaltotal2
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
data = std.fit_transform(df_train)
cols=df_train.columns
df_final = pd.DataFrame(data,columns=cols)
# df_final["Kelas Pekerja"] = df["Kelas Pekerja"]
# df_final["Pendidikan"] = df["Pendidikan"]
# df_final["Status Perkawinan"] = df["Status Perkawinan"]
# df_final["Pekerjaan"] = df["Pekerjaan"]
# df_final["Jenis Kelamin"] = df["Jenis Kelamin"]
std1 = StandardScaler()
data1 = std1.fit_transform(df_test)
cols=df_test.columns
df_test = pd.DataFrame(data1,columns=cols)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
df_drop = df_final.drop(["Umur","Berat Akhir"], axis=1)
df_dropt = df_test.drop(["Umur","Berat Akhir"], axis=1)
```

```python
model = KNeighborsClassifier()
param = {"n_neighbors":np.arange(4,10),"weights":["distance"]}
gscv = GridSearchCV(model,param,scoring="roc_auc",cv=5)
i_model1 = gscv.fit(df_drop,df_target)
```
